```python
import numpy as np
import pandas as pd
df = pd.read_csv('/kaggle/input/10000-books-and-their-genres-standardized/books_and_genres.csv')
df = df.drop(columns=['text'])
```

```python
#search for book based on the book name
from rapidfuzz import process
import ast

# Convert genre strings to actual sets
df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
user_input=input("Enter your preferred book name ")
titles = df['title'].tolist()
best_match, score, index = process.extractOne(user_input, titles)
print(f"ðŸ§  Did you mean: {best_match}? (Confidence: {score:.2f}%)")
matched_book = df.iloc[index]
print(f"\nTitle: {matched_book['title']}")
print(f"Genres: {matched_book['genres']}\n")
target_genres = df.loc[index, 'genres']

# Compute Jaccard similarity with every other book
def jaccard_similarity(set1, set2):
    if not set1 or not set2:
        return 0
    return len(set1 & set2) / len(set1 | set2)

# Store similarities
df['similarity'] = df['genres'].apply(lambda x: jaccard_similarity(target_genres, x))

# Sort by similarity (excluding the book itself)
similar_books = df[df.index != index].sort_values(by='similarity', ascending=False).head(3)

print(f"Since you enjoyed {matched_book['title']} we recommend you these three Books\n")
print(similar_books['title'].to_list())
print("\n")
print("Do you want an explaination?(y/n)\n")
user_input=input()
if(user_input=='y'):
    print("By matching similarity of the genres of your preferred book genres the closes books we found to match your books were\n")
    print(similar_books[['title', 'genres', 'similarity']].head(5))
else:
    print("thanks for using our program\n")
    
    
    
```

```python
raw_input = input("Enter genres (comma-separated): ")
user_genres = set(map(str.strip, raw_input.lower().split(',')))
known_genres = set(g for genres in df['genres'] for g in genres)
user_genres = {process.extractOne(g, known_genres)[0] for g in user_genres}
df['similarity'] = df['genres'].apply(lambda g: jaccard_similarity(user_genres, g))
top_matches = df.sort_values(by='similarity', ascending=False).head(3)
print(f"Since you like the genres {user_genres} we recommend you the following books\n")
print(top_matches['title'].to_list())
print("\n")
print("Do you want an explaination?(y/n)\n")
user_input=input()
if(user_input=='y'):
    print("By matching similarity of the genres of your preferred book genres the closes books we found to match your books were\n")
    print(top_matches[['title', 'genres', 'similarity']].head(5))
else:
    print("thanks for using our program\n")
```
